```
# motionsense_feature_engineering.py
#
# This script takes the combined MotionSense dataset and engineers features
# suitable for anomaly detection models. It uses a sliding window approach
# to create statistical features from the raw sensor data.
#
# How to Use in Google Colab:
# 1. Ensure you have already run the cells to load the MotionSense data
#    into the 'df_combined' DataFrame.
# 2. Place this code in a new Colab cell.
# 3. Run the cell. It will generate a 'motion_features.csv' file, which is
#    your training dataset.

import pandas as pd
import numpy as np
from scipy.stats import kurtosis, skew
import os

# --- Configuration ---
# The DataFrame loaded from the previous cells
# Make sure this variable name matches your loaded data
# input_df is already defined from previous steps

OUTPUT_FEATURES_FILE = 'motion_features.csv'

# Sliding Window Parameters
# Window size: number of sensor readings in one window.
# A size of 100-200 is a good starting point for 50Hz data.
WINDOW_SIZE = 128
# Step size: how many readings to slide the window forward.
# A smaller step creates more overlapping windows and a larger dataset.
STEP_SIZE = 64

# --- Step 1: Isolate a Single User and Activity for Training ---
# We will train our model on what "normal" behavior looks like.
# Let's define "normal" as User 15 performing activity 'A'. You can change this.
# The dataset contains activities: dws, ups, wlk, jog, sit, std
# The dataset contains users 1 through 24.

TARGET_USER = 15
TARGET_ACTIVITY = 'wlk' # Walking

print("Filtering data for User #{0} performing activity '{1}'...".format(TARGET_USER, TARGET_ACTIVITY))

# Recreate the dataframe with user and activity columns using a more robust approach
dataframes = []
for file_path in csv_files:
    try:
        # Extract activity and user from the file path more reliably
        # Example path: 'A_DeviceMotion_data/wlk_15/sub_24.csv'
        parts = file_path.split(os.sep)
        # Assuming activity is in the second-to-last part (e.g., 'wlk_15')
        activity_part = parts[-2]
        # Assuming user is in the last part (e.g., 'sub_24.csv')
        user_part = parts[-1]

        activity = activity_part.split('_')[0]
        user = int(user_part.split('_')[1].split('.')[0])

        df = pd.read_csv(file_path)
        df['user'] = user
        df['activity'] = activity
        dataframes.append(df)
    except (IndexError, ValueError) as e:
        print(f"Could not extract user/activity from file path: {file_path} - {e}")
        continue # Skip this file if extraction fails

# Check if any dataframes were successfully loaded
if not dataframes:
    print("\nError: No dataframes were successfully loaded after attempting to extract user and activity information.")
    features_df = pd.DataFrame() # Ensure features_df is empty
else:
    input_df = pd.concat(dataframes, ignore_index=True)

    # Display unique users and activities AFTER creating input_df
    print("\nUnique users in input_df:", input_df['user'].unique())
    print("Unique activities in input_df:", input_df['activity'].unique())


    # Now, filter the dataframe
    user_activity_df = input_df[
        (input_df['user'] == TARGET_USER) &
        (input_df['activity'] == TARGET_ACTIVITY)
    ].copy()

    # Add a check here to see if the filtered DataFrame is empty
    if user_activity_df.empty:
        print("\nWARNING: Filtered data for User #{0} performing activity '{1}' is empty.".format(TARGET_USER, TARGET_ACTIVITY))
        print("Please check if this user and activity exist in your dataset.")
        # We will skip feature engineering if there's no data
        features_df = pd.DataFrame() # Create an empty DataFrame to avoid errors later
    else:
        # We only need the sensor columns for feature extraction
        sensor_columns = [
            'attitude.roll', 'attitude.pitch', 'attitude.yaw',
            'gravity.x', 'gravity.y', 'gravity.z',
            'rotationRate.x', 'rotationRate.y', 'rotationRate.z',
            'userAcceleration.x', 'userAcceleration.y', 'userAcceleration.z'
        ]
        sensor_df = user_activity_df[sensor_columns]

        print("Filtered data shape: {0}".format(sensor_df.shape))


        # --- Step 2: Create Sliding Windows and Extract Features ---
        print("Creating sliding windows of size {0} with step {1}...".format(WINDOW_SIZE, STEP_SIZE))

        feature_windows = []
        num_samples = len(sensor_df)

        # Iterate through the data with the defined window and step size
        for i in range(0, num_samples - WINDOW_SIZE, STEP_SIZE):
            window = sensor_df.iloc[i : i + WINDOW_SIZE]

            # This will hold the features for the current window
            window_features = {}

            # Calculate features for each sensor column in the window
            for col in sensor_columns:
                # Basic statistics
                window_features['{0}_mean'.format(col)] = window[col].mean()
                window_features['{0}_std'.format(col)] = window[col].std()
                window_features['{0}_min'.format(col)] = window[col].min()
                window_features['{0}_max'.format(col)] = window[col].max()
                window_features['{0}_range'.format(col)] = window[col].max() - window[col].min()
                window_features['{0}_median'.format(col)] = window[col].median()

                # Advanced statistics
                window_features['{0}_skew'.format(col)] = skew(window[col])
                window_features['{0}_kurtosis'.format(col)] = kurtosis(window[col])

            feature_windows.append(window_features)

        # Convert the list of feature dictionaries into a DataFrame
        features_df = pd.DataFrame(feature_windows)

        print("Created {0} feature windows (samples).".format(len(features_df)))


# --- Step 3: Save the Final Feature Set ---
# Only save if the features_df is not empty
if not features_df.empty:
    features_df.to_csv(OUTPUT_FEATURES_FILE, index=False)

    print("-" * 30)
    print("MotionSense feature engineering complete!")
    print("Final feature set saved to '{0}'.".format(OUTPUT_FEATURES_FILE))
    print("Dataset has {0} rows and {1} columns (features).".format(features_df.shape[0], features_df.shape[1]))
    print("-" * 30)
else:
    print("-" * 30)
    print("Feature engineering skipped because no data was filtered.")
    print("-" * 30)
```

```
# model_training.py
#
# This script loads the engineered features from 'motion_features.csv',
# preprocesses the data, and then trains three different anomaly
# detection models: Isolation Forest, a standard Autoencoder, and an
# LSTM Autoencoder. The trained models are saved to disk for later use.
#
# How to Use in Google Colab:
# 1. Ensure you have already run the feature engineering script to generate
#    'motion_features.csv'.
# 2. Place this code in a new Colab cell.
# 3. Run the cell. It will train the models and save them as files in your
#    Colab environment (e.g., 'isolation_forest.joblib', 'autoencoder.h5').

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import IsolationForest
import joblib
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, RepeatVector, TimeDistributed

print("TensorFlow Version: {0}".format(tf.__version__))

# --- Configuration ---
FEATURES_FILE = 'motion_features.csv'

# --- Step 1: Load and Preprocess the Feature Data ---
print("Loading feature data from '{0}'...".format(FEATURES_FILE))
try:
    features_df = pd.read_csv(FEATURES_FILE)
except FileNotFoundError:
    print("Error: The file '{0}' was not found.".format(FEATURES_FILE))
    print("Please make sure you have run the feature engineering script first.")
    exit()

# Add a check to see if the DataFrame is empty
if features_df.empty:
    print("\nError: The loaded feature DataFrame from '{0}' is empty.".format(FEATURES_FILE))
    print("Please ensure the feature engineering script (previous cell) generated features correctly.")
    # Exit gracefully if no data is loaded
    exit()


# For anomaly detection, we don't need a y-label. We only train on normal data.
X_train = features_df.values

# Scale the data to be between 0 and 1. This is crucial for neural networks.
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Save the scaler object. We will need the *exact same* scaler to process
# live data during real-time detection.
joblib.dump(scaler, 'scaler.joblib')
print("Data loaded and scaled. Shape: {0}".format(X_train_scaled.shape))
print("Scaler saved to 'scaler.joblib'.")


# --- Step 2: Train the Isolation Forest Model ---
print("\n--- Training Isolation Forest ---")
# Isolation Forest is great for this task as it's designed for outlier detection.
# 'contamination' is the expected proportion of anomalies in the dataset.
# Since we are training only on "normal" data, we can set it to a small value
# or 'auto'.
iso_forest = IsolationForest(n_estimators=100, contamination='auto', random_state=42)
iso_forest.fit(X_train_scaled)

# Save the trained model to a file
joblib.dump(iso_forest, 'isolation_forest.joblib')
print("Isolation Forest model trained and saved to 'isolation_forest.joblib'.")


# --- Step 3: Train the Standard Autoencoder Model ---
print("\n--- Training Standard Autoencoder ---")
# This model learns to reconstruct the input. High reconstruction error = anomaly.
n_features = X_train_scaled.shape[1]
encoding_dim = 32 # The compressed representation size. Can be tuned.

# Define the model architecture
input_layer = Input(shape=(n_features,))
# Encoder
encoder = Dense(128, activation='relu')(input_layer)
encoder = Dense(64, activation='relu')(encoder)
encoder = Dense(encoding_dim, activation='relu')(encoder)
# Decoder
decoder = Dense(64, activation='relu')(encoder)
decoder = Dense(128, activation='relu')(decoder)
decoder = Dense(n_features, activation='sigmoid')(decoder) # Sigmoid for output between 0-1

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# Print model summary
autoencoder.summary()

# Train the model
autoencoder.fit(
    X_train_scaled,
    X_train_scaled, # The model learns to reconstruct its own input
    epochs=50,
    batch_size=32,
    shuffle=True,
    validation_split=0.1, # Use 10% of data for validation
    verbose=1
)

# Save the trained autoencoder
autoencoder.save('autoencoder.h5')
print("Standard Autoencoder model trained and saved to 'autoencoder.h5'.")


# --- Step 4: Prepare Data and Train the LSTM Autoencoder ---
print("\n--- Training LSTM Autoencoder ---")
# LSTMs are for sequences. We need to reshape our data from 2D to 3D.
# The required shape is [samples, timesteps, features].
# We'll create sequences of 10 consecutive time windows.

def create_sequences(data, time_steps=10):
    sequences = []
    for i in range(len(data) - time_steps + 1):
        sequences.append(data[i : (i + time_steps)])
    return np.array(sequences)

# Define the number of time steps in each sequence
TIME_STEPS = 10
X_train_sequences = create_sequences(X_train_scaled, TIME_STEPS)

if X_train_sequences.shape[0] == 0:
    print("Not enough data to create LSTM sequences with TIME_STEPS={0}. Skipping LSTM training.".format(TIME_STEPS))
else:
    print("Data reshaped for LSTM. New shape: {0}".format(X_train_sequences.shape))
    # Shape is now [num_sequences, TIME_STEPS, n_features]

    # Define the LSTM Autoencoder architecture
    n_features_lstm = X_train_sequences.shape[2]

    lstm_input = Input(shape=(TIME_STEPS, n_features_lstm))
    # Encoder
    encoder_lstm = LSTM(128, activation='relu', return_sequences=False)(lstm_input)
    # This layer repeats the encoded vector TIME_STEPS times
    repeat_vector = RepeatVector(TIME_STEPS)(encoder_lstm)
    # Decoder
    decoder_lstm = LSTM(128, activation='relu', return_sequences=True)(repeat_vector)
    # TimeDistributed applies a Dense layer to each time step of the sequence
    output_lstm = TimeDistributed(Dense(n_features_lstm, activation='sigmoid'))(decoder_lstm)

    lstm_autoencoder = Model(inputs=lstm_input, outputs=output_lstm)
    lstm_autoencoder.compile(optimizer='adam', loss='mean_squared_error')

    # Print model summary
    lstm_autoencoder.summary()

    # Train the LSTM model
    lstm_autoencoder.fit(
        X_train_sequences,
        X_train_sequences, # The model learns to reconstruct the sequence
        epochs=50,
        batch_size=32,
        shuffle=True,
        validation_split=0.1,
        verbose=1
    )

    # Save the trained LSTM autoencoder
    lstm_autoencoder.save('lstm_autoencoder.h5')
    print("LSTM Autoencoder model trained and saved to 'lstm_autoencoder.h5'.")

print("\n--- All models trained successfully! ---")
```

```
# real_time_detection.py
#
# This script simulates the real-time detection process. It loads the
# trained models and the scaler, then processes new data (from a different
# user or activity) to see if the models can detect it as an anomaly.
#
# How to Use in Google Colab:
# 1. Ensure you have run the previous scripts to generate:
#    - 'motion_features.csv'
#    - 'scaler.joblib'
#    - 'isolation_forest.joblib'
#    - 'autoencoder.h5'
#    - 'lstm_autoencoder.h5'
# 2. You also need the original combined dataframe ('df_combined' or similar)
#    from the data loading step to simulate new, unseen data.
# 3. Place this code in a new Colab cell and run it.

import pandas as pd
import numpy as np
import joblib
import tensorflow as tf
from scipy.stats import kurtosis, skew
import os # Import os for path splitting

# --- Configuration ---
# This should be the same configuration as your feature engineering script
WINDOW_SIZE = 128
STEP_SIZE = 64
TIME_STEPS = 10 # For the LSTM model

# Anomaly Threshold for Autoencoders. This is a critical parameter.
# It's the Mean Squared Error (MSE) above which we consider a sample anomalous.
# This value often needs to be tuned based on a validation set. We'll start
# with a value determined by analyzing the training data's reconstruction error.
AUTOENCODER_THRESHOLD = 0.01 # You may need to adjust this value
LSTM_THRESHOLD = 0.01        # You may need to adjust this value

# --- Step 1: Load Models and Scaler ---
print("--- Loading all trained models and the scaler ---")
try:
    scaler = joblib.load('scaler.joblib')
    iso_forest = joblib.load('isolation_forest.joblib')
    # Load models with compile=False as we are only doing inference
    autoencoder = tf.keras.models.load_model('autoencoder.h5', compile=False)
    lstm_autoencoder = tf.keras.models.load_model('lstm_autoencoder.h5', compile=False)
    print("Models and scaler loaded successfully.")
except FileNotFoundError as e:
    print("Error loading model files: {0}".format(e))
    print("Please ensure all model and scaler files are present in the Colab environment.")
    exit()

# --- Step 2: Prepare Anomalous Data for Simulation ---
# We will now load data from a *different* user or activity to test our system.
# Our models were trained on what was filtered in the previous cell (TARGET_USER, TARGET_ACTIVITY).
# Let's test on a potentially anomalous behavior, e.g., User 23 jogging.

ANOMALOUS_USER = 23
ANOMALOUS_ACTIVITY = 'jog'

print("\n--- Preparing anomalous data (User #{0}, Activity: '{1}') ---".format(ANOMALOUS_USER, ANOMALOUS_ACTIVITY))

# This assumes 'input_df' (the fully combined dataframe with user/activity columns)
# is available from your previous cells. If not, you may need to reload it.
try:
    # Recreate the dataframe with user and activity columns within this script
    # to ensure consistency, using the same logic as the feature engineering script.
    dataframes = []
    for file_path in csv_files: # Assuming csv_files is available from previous steps
        try:
            # Extract activity and user from the file path more reliably
            parts = file_path.split(os.sep)
            activity_part = parts[-2]
            user_part = parts[-1]

            activity = activity_part.split('_')[0]
            user = int(user_part.split('_')[1].split('.')[0])

            df = pd.read_csv(file_path)
            df['user'] = user
            df['activity'] = activity
            dataframes.append(df)
        except (IndexError, ValueError) as e:
            # print(f"Could not extract user/activity from file path: {file_path} - {e}") # Optional: keep for debugging
            continue # Skip this file if extraction fails

    if not dataframes:
         print("\nError: No dataframes were successfully loaded for anomalous data preparation.")
         exit()
    else:
        full_dataset_for_anomaly_test = pd.concat(dataframes, ignore_index=True)


    anomaly_df_raw = full_dataset_for_anomaly_test[
        (full_dataset_for_anomaly_test['user'] == ANOMALOUS_USER) &
        (full_dataset_for_anomaly_test['activity'] == ANOMALOUS_ACTIVITY)
    ].copy()

except NameError:
    print("Error: The 'csv_files' variable is not defined. Please run the data loading and file path identification steps first.")
    exit()
except Exception as e:
    print(f"An error occurred while preparing anomalous data: {e}")
    exit()


# Check if the anomalous data DataFrame is empty
if anomaly_df_raw.empty:
    print("\nWARNING: Filtered data for Anomalous User #{0} and activity '{1}' is empty.".format(ANOMALOUS_USER, ANOMALOUS_ACTIVITY))
    print("Please check if this user and activity exist in your dataset for testing.")
    # If no anomalous data, we can't proceed with detection simulation
    exit()


sensor_columns = [
    'attitude.roll', 'attitude.pitch', 'attitude.yaw', 'gravity.x', 'gravity.y',
    'gravity.z', 'rotationRate.x', 'rotationRate.y', 'rotationRate.z',
    'userAcceleration.x', 'userAcceleration.y', 'userAcceleration.z'
]
anomaly_df_sensors = anomaly_df_raw[sensor_columns]

# --- Step 3: Feature Engineering for the Anomalous Data ---
# We must apply the *exact same* feature engineering process.
feature_windows = []
num_anomaly_samples = len(anomaly_df_sensors)

# Ensure we have enough samples for at least one window
if num_anomaly_samples < WINDOW_SIZE:
    print(f"\nNot enough anomalous data samples ({num_anomaly_samples}) to create a window of size {WINDOW_SIZE}. Skipping detection simulation.")
    exit()


for i in range(0, num_anomaly_samples - WINDOW_SIZE, STEP_SIZE):
    window = anomaly_df_sensors.iloc[i : i + WINDOW_SIZE]
    window_features = {}
    for col in sensor_columns:
        # Basic statistics
        window_features['{0}_mean'.format(col)] = window[col].mean()
        window_features['{0}_std'.format(col)] = window[col].std()
        window_features['{0}_min'.format(col)] = window[col].min()
        window_features['{0}_max'.format(col)] = window[col].max()
        window_features['{0}_range'.format(col)] = window[col].max() - window[col].min()
        window_features['{0}_median'.format(col)] = window[col].median()

        # Advanced statistics
        # Handle cases with insufficient data points for skew/kurtosis (e.g., if window[col].std() is 0)
        if window[col].std() > 0:
             window_features['{0}_skew'.format(col)] = skew(window[col])
             window_features['{0}_kurtosis'.format(col)] = kurtosis(window[col])
        else:
             window_features['{0}_skew'.format(col)] = 0.0
             window_features['{0}_kurtosis'.format(col)] = 0.0

    feature_windows.append(window_features)

anomaly_features_df = pd.DataFrame(feature_windows)

# --- Align columns with the training data features ---
# Load the original feature column order from the training data CSV
try:
    original_features_df_for_cols = pd.read_csv('motion_features.csv')
    training_columns = original_features_df_for_cols.columns.tolist()

    # Reindex the anomaly features DataFrame to match the training columns
    # Fill any missing columns in anomaly_features_df with 0 (or another appropriate value)
    anomaly_features_df = anomaly_features_df.reindex(columns=training_columns, fill_value=0)

except FileNotFoundError:
     print("Error: 'motion_features.csv' not found. Cannot align columns for anomaly detection.")
     exit()
except Exception as e:
     print(f"Error aligning columns: {e}")
     exit()


print("Engineered {0} feature windows from anomalous data.".format(len(anomaly_features_df)))

# --- Step 4: Scale the New Data and Make Predictions ---
# Use the *already fitted* scaler from training. DO NOT fit again.
X_anomaly_scaled = scaler.transform(anomaly_features_df.values)

# Prepare sequences for LSTM
def create_sequences(data, time_steps=10):
    sequences = []
    for i in range(len(data) - time_steps + 1):
        sequences.append(data[i:(i + time_steps)])
    return np.array(sequences)

X_anomaly_sequences = create_sequences(X_anomaly_scaled, TIME_STEPS)

print("\n--- Making Predictions on Anomalous Data ---")

# 1. Isolation Forest Prediction
# predict() returns -1 for outliers (anomalies) and 1 for inliers (normal).
if len(X_anomaly_scaled) > 0:
    iso_preds = iso_forest.predict(X_anomaly_scaled)
    iso_anomaly_count = np.sum(iso_preds == -1)
    print("Isolation Forest found {0}/{1} samples to be ANOMALOUS.".format(iso_anomaly_count, len(iso_preds)))
else:
    print("No scaled anomalous data to predict with Isolation Forest.")
    iso_preds = np.array([]) # Ensure iso_preds is defined


# 2. Standard Autoencoder Prediction
# We calculate the reconstruction error (MSE) for each sample.
if len(X_anomaly_scaled) > 0:
    ae_reconstructions = autoencoder.predict(X_anomaly_scaled)
    ae_mse = np.mean(np.power(X_anomaly_scaled - ae_reconstructions, 2), axis=1)
    ae_anomalies = ae_mse > AUTOENCODER_THRESHOLD
    print("Standard Autoencoder found {0}/{1} samples to be ANOMALOUS (MSE > {2}).".format(np.sum(ae_anomalies), len(ae_anomalies), AUTOENCODER_THRESHOLD))
else:
    print("No scaled anomalous data to predict with Standard Autoencoder.")
    ae_mse = np.array([]) # Ensure ae_mse is defined


# 3. LSTM Autoencoder Prediction
if len(X_anomaly_sequences) > 0:
    lstm_reconstructions = lstm_autoencoder.predict(X_anomaly_sequences)
    # Calculate MSE for each sequence
    lstm_mse = np.mean(np.power(X_anomaly_sequences - lstm_reconstructions, 2), axis=(1, 2))
    lstm_anomalies = lstm_mse > LSTM_THRESHOLD
    print("LSTM Autoencoder found {0}/{1} sequences to be ANOMALOUS (MSE > {2}).".format(np.sum(lstm_anomalies), len(lstm_anomalies), LSTM_THRESHOLD))
else:
    print("Not enough anomalous data sequences for LSTM prediction.")
    lstm_anomalies = np.array([]) # Ensure lstm_anomalies is defined


# --- Step 5: Decision Engine (Example) ---
# In a real application, you would combine these results.
# A simple majority vote is a good starting point.
print("\n--- Final System Decision Simulation ---")

# Ensure we have predictions for the samples we want to check
max_samples_to_check = min(5, len(X_anomaly_scaled))

if max_samples_to_check > 0:
    # Iterate through the first few samples or available samples
    for i in range(max_samples_to_check):
        # Get votes from each model for this sample
        iso_vote = 1 if len(iso_preds) > i and iso_preds[i] == -1 else 0
        ae_vote = 1 if len(ae_mse) > i and ae_mse[i] > AUTOENCODER_THRESHOLD else 0
        # For LSTM, we check if the sequence *containing* this sample is anomalous
        # This is a simplification; a proper implementation might need to map
        # LSTM sequence anomalies back to individual samples.
        lstm_vote = 0
        # Check if the sample index is within the range of LSTM sequences
        # A sample at index 'i' is part of LSTM sequences from max(0, i - TIME_STEPS + 1) to i
        # For simplicity here, we'll check if the sequence *starting* at 'i' is anomalous,
        # but a more robust check would be needed for overlapping windows.
        # Let's just check if *any* LSTM anomaly occurred up to this point for simplicity in the demo.
        # A better approach would involve aligning LSTM output to feature windows.
        # For this demo, let's just check if the sample index aligns with an LSTM sequence start
        lstm_seq_index = i // STEP_SIZE # Simple mapping, may not be accurate with overlapping windows
        if len(lstm_anomalies) > lstm_seq_index and lstm_anomalies[lstm_seq_index]:
             lstm_vote = 1


        total_votes = iso_vote + ae_vote + lstm_vote
        decision = "ANOMALY - LOCK SESSION" if total_votes >= 2 else "Normal - Allow"

        print("Sample #{0}: Votes=[Iso:{1}, AE:{2}, LSTM:{3}] -> Total: {4} -> Decision: {5}".format(
            i, iso_vote, ae_vote, lstm_vote, total_votes, decision
        ))
else:
    print("No anomalous data samples available for decision simulation.")

print("\n--- Anomaly detection simulation complete! ---")
```
